```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
d=randn(25)
```

```python
sns.rugplot(d)
```

```python
plt.hist(d,alpha=0.3)
sns.rugplot(d)
```

```python
sns.rugplot(d)
x_min = d.min() -2
x_max = d.max() + 2

x_axis = np.linspace(x_min,x_max,100)
bandwidth =  ((4*d.std()**5) / (3*len(d))) ** 0.2  

kl= []

for dpoint in d:
    kernel= stats.norm(dpoint,bandwidth).pdf(x_axis)
    kl.append(kernel)
    
    kernel = kernel /kernel.max()
    kernel=kernel*0.4
    plt.plot(x_axis,kernel,color='grey',alpha=0.5)
    
plt.ylim(0,1)
```

```python
sns.rugplot(d)
```

```python
kl
```

```python
sumd = np.sum(kl,axis=0)
fig=plt.plot(x_axis,sumd,color='indianred')
sns.rugplot(d)
#plt.yticks([])
plt.suptitle("Sum of basis Functions")
```
