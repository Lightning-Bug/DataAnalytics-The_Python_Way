```python
# The normal imports
import pandas as pd
import numpy as np
from numpy.random import randn

# Import the stats library from numpy
from scipy import stats

# Import plotting modules
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots appear in iPython notebook
%matplotlib inline
```

```python
tip =sns.load_dataset('tips')
tip.head()
```

```python
sns.lmplot('total_bill','tip',tip)
```

```python
sns.lmplot('total_bill','tip',tip,order=4,scatter_kws={'marker':'o','color':'indianred'},
                    line_kws={'linewidth':1,'color':'blue'})
```

```python
tip.head()
```

```python
tip['perc']=100*tip['tip']/tip['total_bill']
tip.head()
```

```python
sns.lmplot('size','perc',tip)
```

```python
sns.lmplot('size','perc',tip,x_jitter=1)
```

```python
sns.lmplot('size','perc',tip,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','perc',tip,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','perc',tip,hue='day')
```

```python
sns.lmplot('total_bill','perc',tip,lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot('total_bill','perc',tip)
```

```python
fig, (axis1,axis2)=plt.subplots(1,2,sharey=True)
sns.regplot('total_bill','perc',tip,ax=axis1)
sns.violinplot(tip['perc'],tip['size'],hue = 'day',palette='muted',ax=axis2)
```
