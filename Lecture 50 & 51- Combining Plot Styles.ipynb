# Box plots and Lecture 51


```python
d1=randn(100)
d2=randn(100)
```

```python
sns.boxplot([d1,d2])
```

```python
sns.boxplot([d1,d2],whis=np.inf)
```

```python
sns.boxplot(d1, hor=True)
```

```python
d1=stats.norm(0,5).rvs(100)

d2= np.concatenate([stats.gamma(5).rvs(50)-1,-1*stats.gamma(5).rvs(50)])

sns.boxplot([d1,d2],whis=np.inf)
```

```python
sns.violinplot([d1,d2])
```

```python
sns.violinplot([d1,d2],bw=0.01)
```

```python
sns.violinplot([d1])
```
